```python
import numpy as np

a = np.array([0, 1, 2])
print(type(a))
```

```python
# Create a rank 2 array:
b = np.array([[0,1,2],[3,4,5]])
print(b.shape)
print(b)
print(b[0, 0], b[0, 1], b[1, 0])
```

```python
# Create a 3x3 array of all zeros:
a = np.zeros((3,3))
print(a)
```

```python
# Create a 2x2 array of all ones:
b = np.ones((2,2))
print(b)
```

```python
# Create a 3x3 constant array:
c = np.full((3,3), 7)
print(c)
```

```python
# Create a 3x3 array filled with random values:
d = np.random.random((3,3))
print (d)
```

```python
# Create a 3x3 identity matrix:
e = np.eye(3)
print (e)
```

```python
# mix of tuple and lists
h = np.array([[0, 1,2.0],[0,0,0],(1+1j,3.,2.)])
print (h)
```

```python
j = np.linspace(2., 4., 7)
print (j)
```

```python
# linspace() will create arrays with a specified number of items which are spaced equally between the specified beginning and end values:
k = np.indices((3,2))
print (k)
```

```python
# indices() will create a set of arrays stacked as a one-higher dimensioned array, one per dimension with each representing variation in that dimension.
k = np.indices((2,2))
print (k)
```

```python
# basic slicing
x = np.array([5, 6, 7, 8, 9])
print(x[1:7:2])
```

```python
y = np.array([[[1],[2],[3]], [[4],[5],[6]]])
print ("Shape of y: ", y.shape, "\n", y[1:3])
```

```python
a = np.array([[5,6,7,8], [1,2,3,4], [9,10,11,12]])
b = a[:2, 1:3]
print ("Array b:", b)
```

```python
# boolean
a=np.array([[1,2], [3, 4], [5, 6]])
print (a > 2)
print (a[a>2])
```

```python
# array math
x=np.array([[1,2],[3,4],[5,6]])
y=np.array([[7,8],[9,10],[11,12]])
print(x+y)
print(np.add(x, y))
print(x-y)
print(np.subtract(x, y))
```
