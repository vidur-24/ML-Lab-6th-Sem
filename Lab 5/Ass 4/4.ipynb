## Q4
Write a program that:

- Trains linear regression using gradient descent  
- Stops training if cost increases for 5 consecutive iterations  

### Test with:
- α = 0.01  
- α = 1.0  

### Print:
- Whether the model converged or diverged  

```python
# Cell 1: Import libraries
import numpy as np
```

```python
# Cell 2: Define dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
m = len(X)
```

```python
# Cell 3: Gradient Descent with early stopping (5 consecutive increases)

def gradient_descent_early_stop(X, y, alpha, iterations):
    w = 0
    b = 0
    
    prev_cost = float('inf')
    increase_count = 0

    for i in range(iterations):
        y_pred = w * X + b
        
        cost = (1/(2*m)) * np.sum((y_pred - y)**2)
        
        # Check if cost increased
        if cost > prev_cost:
            increase_count += 1
        else:
            increase_count = 0
        
        # Stop if cost increases 5 times consecutively
        if increase_count >= 5:
            return w, b, cost, "Diverged"
        
        prev_cost = cost
        
        # Gradients
        dw = (1/m) * np.sum((y_pred - y) * X)
        db = (1/m) * np.sum(y_pred - y)
        
        # Update
        w -= alpha * dw
        b -= alpha * db

    return w, b, cost, "Converged"
```

```python
# Cell 4: Test for alpha = 0.01 and 1.0

learning_rates = [0.01, 1.0]
iterations = 1000

for alpha in learning_rates:
    w, b, cost, status = gradient_descent_early_stop(X, y, alpha, iterations)
    
    print(f"\nAlpha = {alpha}")
    print(f"Final w = {w}")
    print(f"Final b = {b}")
    print(f"Final Cost = {cost}")
    print(f"Status: {status}")
```

### Results and Observations

- **For α = 0.01:**  
  The cost decreases consistently over iterations, and the model successfully converges to optimal parameter values. No early stopping is triggered, indicating stable learning.

- **For α = 1.0:**  
  The cost increases rapidly over consecutive iterations, and early stopping is triggered after 5 consecutive increases. This indicates that the model diverges due to excessively large parameter updates.

### Observation

The experiment shows that:
- A **moderate learning rate (α = 0.01)** ensures stable and gradual convergence.
- A **large learning rate (α = 1.0)** causes overshooting of the minimum, leading to divergence.
- Early stopping effectively prevents unnecessary computations when the model fails to converge.
