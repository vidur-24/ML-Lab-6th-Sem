##### Q. To perform univariate data analysis using graphical visualization techniques including count plots, pie charts, histograms, and box plots and to identify and analyse outliers within the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../supervised (gender class)/gender_classification_v7.csv")

# View first rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
binary_cols = ['long_hair','nose_wide','nose_long',
               'lips_thin','distance_nose_to_lip_long']

for col in binary_cols:
    plt.figure()
    sns.countplot(x=col, data=df)
    plt.title(f"Count Plot of {col}")
    plt.show()
```

```python
df['gender'].value_counts().plot.pie(
    autopct='%1.1f%%',
    figsize=(5,5)
)
plt.title("Gender Proportion")
plt.ylabel("")
plt.show()
```

```python
sns.histplot(df['forehead_width_cm'], bins=8)
plt.title("Histogram of Forehead Width")
plt.show()

sns.histplot(df['forehead_height_cm'], bins=8)
plt.title("Histogram of Forehead Height")
plt.show()
```

```python
sns.boxplot(x=df['forehead_width_cm'])
plt.title("Box Plot of Forehead Width")
plt.show()

sns.boxplot(x=df['forehead_height_cm'])
plt.title("Box Plot of Forehead Height")
plt.show()
```

```python
for col in ['forehead_width_cm','forehead_height_cm']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}")
    print(f"Lower Bound: {lower}")
    print(f"Upper Bound: {upper}")
    print(f"Number of Outliers: {len(outliers)}\n")
```
