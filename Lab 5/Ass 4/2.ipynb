## Q2
Write a Python program to implement Linear Regression using Gradient Descent.

### Requirements:
- Dataset:
  - X = [1, 2, 3, 4, 5]
  - y = [2, 4, 6, 8, 10]

- Train the model using learning rates:
  - α = 0.001, 0.01, 0.1

- Run for 1000 iterations.

### Print:
- Final weight and bias  
- Final cost  

### Task:
- Compare convergence speed.

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

m = len(X)

def gradient_descent(X, y, alpha, iterations):
    w = 0
    b = 0

    for i in range(iterations):
        y_pred = w * X + b
        
        dw = (1/m) * np.sum((y_pred - y) * X)
        db = (1/m) * np.sum(y_pred - y)
        
        w = w - alpha * dw
        b = b - alpha * db

    cost = (1/(2*m)) * np.sum((w*X + b - y)**2)
    
    return w, b, cost


learning_rates = [0.001, 0.01, 0.1]

for alpha in learning_rates:
    w, b, cost = gradient_descent(X, y, alpha, 1000)
    
    print(f"\nLearning Rate: {alpha}")
    print(f"Final Weight (w): {w}")
    print(f"Final Bias (b): {b}")
    print(f"Final Cost: {cost}")
```

### Comparison of Convergence Speed

- **α = 0.001:**  
  Converges **very slowly**. The cost decreases gradually and requires many iterations to approach the optimal solution.

- **α = 0.01:**  
  Converges at a **moderate speed**. The cost reduces efficiently and reaches near-optimal values within the given iterations.

- **α = 0.1:**  
  Converges **fastest**. The cost drops rapidly and the model reaches the optimal solution in fewer iterations.

### Conclusion

The convergence speed increases with the learning rate:
- **Slowest:** α = 0.001  
- **Moderate:** α = 0.01  
- **Fastest:** α = 0.1  

Thus, **α = 0.1 provides the fastest convergence** for this dataset.
