# Q1 — Load Product Review Dataset

In this question, we load the Amazon product reviews dataset using pandas.
The dataset contains product reviews and ratings which will be used for sentiment analysis.

```python
# Cell 1: Load Product Review Dataset

import pandas as pd

# Load dataset
df = pd.read_csv("../../dataset/amazon_review.csv")

# Select required columns
df = df[['reviewText', 'overall']]

# Rename columns
df.columns = ['Review', 'Rating']

print(df.head())
```

# Q2 — Text Preprocessing

In this step, the review text is cleaned using Natural Language Processing techniques:
- Convert text to lowercase
- Remove punctuation
- Remove stopwords
- Perform tokenization
- Perform stemming

The cleaned text is stored in a new column.

```python
# Cell 2: Text Preprocessing

import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

stemmer = PorterStemmer()

def preprocess_text(text):

    text = str(text)

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    # Tokenization
    tokens = word_tokenize(text)

    # Remove stopwords and stemming
    cleaned_tokens = []

    for word in tokens:

        if word not in stop_words:

            cleaned_tokens.append(
                stemmer.stem(word)
            )

    return " ".join(cleaned_tokens)

# Apply preprocessing
df['Cleaned_Review'] = df['Review'].apply(preprocess_text)

print(df[['Review', 'Cleaned_Review']].head())
```

# Q3 — Convert Text into Numerical Features

Machine learning models cannot understand text directly.
Therefore, the cleaned reviews are converted into numerical features using TF-IDF Vectorization.

TF-IDF assigns importance scores to words based on their frequency.

```python
# Cell 3: TF-IDF Feature Extraction

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(
    df['Cleaned_Review']
)

print("Feature Matrix Shape:")

print(X.shape)

print("\nFeature Names:\n")

print(vectorizer.get_feature_names_out())

print("\nFirst 10 Rows of Feature Matrix:\n")

print(X[:10])
```

# Q4 — Create Sentiment Labels

Sentiment labels are created using product ratings.
- Rating greater than or equal to 5 → Positive
- Otherwise → Negative

These labels will be used for classification.

```python
# Cell 4: Create Sentiment Labels

df['Rating'] = df['Rating'].astype(int)

# Rating >= 5 => Positive
# Else => Negative

df['Sentiment'] = df['Rating'].apply(
    lambda x: 'Positive' if x >= 5 else 'Negative'
)

print(df[['Review', 'Rating', 'Sentiment']].head())
```

# Q5 — Train Different Classification Algorithms

Different machine learning classification algorithms are trained on the dataset:
- Logistic Regression
- Naive Bayes
- Decision Tree

The dataset is first divided into training and testing data.

```python
# Cell 5: Train Classification Models

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

# Target variable
y = df['Sentiment']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

print("All Models Trained Successfully")
```

# Q6 — Evaluate Classifier Performance

The performance of each classifier is evaluated using:
- Accuracy
- Precision
- Recall
- F1-Score
- Confusion Matrix

A comparative analysis is also performed.

```python
# Cell 6: Evaluate Models

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

# Predictions
lr_pred = lr_model.predict(X_test)
nb_pred = nb_model.predict(X_test)
dt_pred = dt_model.predict(X_test)

models = {
    "Logistic Regression": lr_pred,
    "Naive Bayes": nb_pred,
    "Decision Tree": dt_pred
}

for model_name, pred in models.items():

    print("\n", model_name)

    print("Accuracy:",
          accuracy_score(y_test, pred))

    print("Precision:",
          precision_score(
              y_test,
              pred,
              pos_label='Positive'
          ))

    print("Recall:",
          recall_score(
              y_test,
              pred,
              pos_label='Positive'
          ))

    print("F1 Score:",
          f1_score(
              y_test,
              pred,
              pos_label='Positive'
          ))

    print("\nConfusion Matrix:\n")

    print(confusion_matrix(y_test, pred))
```

# Q7 — Predict Sentiment of New Review

A prediction function is created to classify a new product review as:
- Positive
- Negative

The trained Logistic Regression model is used for prediction.

```python
# Cell 7: Predict Sentiment of New Review

def predict_sentiment(review):

    # Preprocess review
    cleaned_review = preprocess_text(review)

    # Convert into vector
    review_vector = vectorizer.transform(
        [cleaned_review]
    )

    # Predict sentiment
    prediction = lr_model.predict(review_vector)

    return prediction[0]

# Test prediction
test_review = "This product is amazing and very useful"

result = predict_sentiment(test_review)

print("Review:", test_review)

print("Predicted Sentiment:", result)
```
